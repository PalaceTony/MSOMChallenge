```python
import pandas as pd

fd1_fd2 = pd.read_csv("data/processed/fd1_fd2.csv")
fd1_fd2
```

```python
fd1_fd2["TimeStamp"] = pd.to_datetime(fd1_fd2["TimeStamp"])
fd1_fd2.dtypes
```

# fd1_fd2 Columns


```python
fd1_fd2.columns
```

```python
fd1_fd2
```

# fd1_fd2_aggregated

```python
fd1_fd2_aggregated = (
    fd1_fd2.groupby(fd1_fd2["TimeStamp"].dt.floor("S"))
    .mean(numeric_only=True)
    .reset_index()
)


fd1_fd2_aggregated
```

# Blenders

```python
import pandas as pd

blenders = pd.read_csv("data/processed/blenders.csv")
blenders["TimeStamp"] = pd.to_datetime(blenders["TimeStamp"])
blenders
```

```python
blenders.dtypes
```

# Blenders aggregated

```python
blenders_aggregated = (
    blenders.groupby(blenders["TimeStamp"].dt.floor("S"))
    .mean(numeric_only=True)
    .reset_index()
)


blenders_aggregated
```

# Visualize the data

## Feedfactor 1-7

```python
import matplotlib.pyplot as plt

for i in range(1, 8):
    if i == 6:  # Skipping PD 6
        continue

    # Creating a new figure for each PD mass flow
    plt.figure(figsize=(10, 6))
    plt.scatter(
        fd1_fd2_aggregated["TimeStamp"],
        fd1_fd2_aggregated[f"Feed Factor PD{i}"],
        color="b",
        s=3,  # Adjust the 's' parameter for point size
    )
    plt.title(f"Feed Factor PD{i} Over Time")
    plt.xlabel("Timestamp")
    plt.ylabel(f"Feed Factor PD{i}")
    plt.grid(True)

    # Formatting the x-ticks for better readability
    plt.xticks(rotation=45)
    plt.gcf().autofmt_xdate()

    # Adjust layout
    plt.tight_layout()
    plt.show()
```
