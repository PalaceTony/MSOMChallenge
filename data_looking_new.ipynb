```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

unique_lots_dict = {
    1: fd1_fd2["PD1Lot"].unique(),
    2: fd1_fd2["PD2Lot"].unique(),
    3: fd1_fd2["PD3Lot"].unique(),
    4: fd1_fd2["PD4Lot"].unique(),
    5: fd1_fd2["PD5Lot"].unique(),
    7: fd1_fd2["PD7Lot"].unique(),
}

for i in range(1, 8):
    if i == 6:  # Skipping PD 6
        continue

    # Select unique lots for the current PD
    unique_lots = unique_lots_dict[i]

    # Use a more distinct color palette
    colors = plt.get_cmap("tab20")(np.linspace(0, 1, len(unique_lots)))

    # Creating a new figure for each PD mass flow
    plt.figure(figsize=(12, 8))

    # Loop through each lot and plot points with corresponding color
    for idx, lot in enumerate(unique_lots):
        lot_mask = (
            fd1_fd2[f"PD{i}Lot"] == lot
        )  # Correctly choose the current PD lot column

        plt.scatter(
            fd1_fd2["TimeStamp"][lot_mask],
            fd1_fd2[f"Feed Factor PD{i}"][lot_mask],
            color=colors[idx % len(colors)],  # Use distinct colors for each lot
            s=8,  # Increase the 's' parameter for larger point size
            alpha=0.9,  # Slightly decrease transparency for clearer view
        )

    # Adding labels and title
    plt.title(f"Feed Factor PD{i} Over Time with Lot Changes")
    plt.xlabel("Timestamp")
    plt.ylabel(f"Feed Factor PD{i}")
    plt.grid(False)

    # Simplify the legend explanation
    plt.text(
        0.95,
        0.95,
        "Colors represent different lots",
        verticalalignment="top",
        horizontalalignment="right",
        transform=plt.gca().transAxes,
        color="black",
        fontsize=12,
    )

    # Formatting the x-ticks for better readability
    plt.xticks(rotation=45)
    plt.gcf().autofmt_xdate()

    # Adjust layout
    plt.tight_layout()

    # Show plot
    plt.show()
```
